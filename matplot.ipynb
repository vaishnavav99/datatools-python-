```python
#styling and adding features
from matplotlib import style
style.use('ggplot')

x1=[1,2,3]
y1=[2,3,4]
x2=[3,4,5]
y2=[6,5,4]


plt.plot(x1,y1,'g',label='L1',linewidth=4)
plt.plot(x2,y2,'r',label='L2',linewidth=2)

#label,title
plt.title('Graph Styled')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

#display line details
plt.legend()

#try different grid
#plt.grid(True,color='k')
#plt.grid(False)

plt.show()
```

```python
#Bar Graph

x3=[1,3,5]
y3=[10,8,14]
x4=[2,4,6]
y4=[22,15,9]

plt.bar(x3,y3,color='g',label='B1')
plt.bar(x4,y4,color='c',label='B2')

#label,title
plt.title('Bar Graph ')
plt.xlabel('number')
plt.ylabel('height')

#display line details
plt.legend()


plt.show()
```

```python
#Histrogram

population_ages=[10,20,50,70,80,60,90]
bins=[10,18,30,65,100]


plt.hist(population_ages,bins,histtype='bar',rwidth=0.8)


#label,title
plt.title('population chart ')
plt.xlabel('age')
plt.ylabel('population')

plt.show()
```

```python
#pie plot
slices=[7,5,2,15]

activities=['sleep','work','eat','play']
cols=['c','b','g','y']
#startangle use to specify start
plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0,0.1,0),autopct='%1.1f%%')

plt.title('pie chart')
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[3,5,6,2,3,5,7,9]

plt.scatter(x,y,label='dot',color='k',s=25)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(False)
plt.show()
```

```python
day=[1,2,3,4,5]
sleep= [10,9,7,8,8]
eat= [2,3,4,2,3]
play= [12,10,6,8,9]
work= [0,2,7,6,4]

plt.plot([],[],color='m',label='sleep',linewidth=5)
plt.plot([],[],color='c',label='eat',linewidth=5)
plt.plot([],[],color='r',label='play',linewidth=5)
plt.plot([],[],color='k',label='work',linewidth=5)

plt.stackplot(day,sleep,eat,play,work,colors=['m','c','r','k'])

plt.title('Stack Plot')
plt.xlabel('day')
plt.ylabel('hours')
plt.legend()
plt.grid(False)
plt.show()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
